```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import pandas as pd
```

```python
fcpo_data_daily=pd.read_hdf('processed_dta.h5','fcpo_data_daily')
palmoil_data_daily=pd.read_hdf('processed_dta.h5','palmoil_data_daily')
```

Note: The indices of FCPO and palm oil are not the same, hence the direct correlation doesn't work. Need to find the intersection

```python
# selected time ranges
selected_fcpo_data=fcpo_data_daily['2018-01-01':'2018-06-30']
selected_poil_data=palmoil_data_daily['2018-01-01':'2018-06-30']
```

```python

fig,(ax1,ax2) = plt.subplots(2,1)
fig.set_size_inches(20,15)
ax1.plot(selected_fcpo_data['Close'],label='fcpo')
ax1.set_title('FCPO Prices')

ax2.plot(selected_fcpo_data['Close']/selected_fcpo_data['Close'].iloc[0],label='fcpo')
ax2.plot(selected_poil_data['Close']/selected_poil_data['Close'].iloc[0],label='palmoil')
ax2.set_title('Relation between FCPO and Palmoil')
```
